## Import pandas library

```python
import pandas as pd
```

## Load the data from auto.json file

```python
df = pd.read_json('../data/auto.json', orient='records')
df.set_index(['CarNumber'], inplace=True)
df
```

## Display rows where fines > 2100

```python
df[df['Fines'] > 2100]
```

## Display rows where Fines > 2100 and Refund = 2

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

## Display rows where Models are in ['Focus', 'Corolla']

```python
df[df['Model'].isin(['Focus', 'Corolla'])]
```

## Display rows where the car numbers are from the list:
['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(car_numbers)]
```

## Display the median fines grouped by the make

```python
df.groupby('Make')['Fines'].median()
```

## Display the median of fines grouped by the make and the model

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

## Display the number of fines grouped by the make and the model

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

## Display the minimum and the maximum fines grouped by the make and the model 

```python
# min_fines = df.groupby(['Make', 'Model'])['Fines'].min()
# max_fines = df.groupby(['Make', 'Model'])['Fines'].max()
# print(f"Min fines:\n{min_fines}\nMax fines:\n{max_fines}")

df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

## Display the standard defiation of the fines grouped by the Make and model

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

# Aggregations with the car number
## Display the car numbers grouped by the number of fines

```python
fines_count_by_car = df.groupby('CarNumber').size().sort_values(ascending=False)
fines_count_by_car
```

## Rows corresponding to th top-1 car number who most often violated the law

```python
top_cars_by_fines_count = fines_count_by_car.index[0]
rows_top_car_fines_count = df[df.index == top_cars_by_fines_count]
rows_top_car_fines_count
```

## Car numbers by sum of the fines(by who paid the most)

```python
fines_sum_by_car =  df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
fines_sum_by_car
```

## Row for top-1 car who paid the most

```python
top_car_by_fines_count = fines_sum_by_car.index[0]
rows_top_car_fines_count = df[df.index == top_car_by_fines_count]
rows_top_car_fines_count
```

## Are there any car numbers that were connected to different models?

```python
car_numbers_multiple_models = df.groupby('CarNumber')['Model'].nunique()
car_numbers_multiple_models = car_numbers_multiple_models[car_numbers_multiple_models > 1]
f"Yes there are {len(car_numbers_multiple_models)} cars" if len(car_numbers_multiple_models) else "No"
```
